# Intro

- Partitioning a customer base into groups of individuals based on characteristics
- Allows a business to target different  groups (high profit&low risk,...)
- we can cross-refrence the groups with their purchases



- finding "clusters" in datasets, unsupervised (unlike supervised, we don't tell them from the start what's what and which is which and so on)
- cluster: a group of data points or objects in a dataset that are similar to other objects in the group, and dissimilar to datapoints in other clusters
- diffrent than classification:
    - no need to be labeled (unsupervised)
    - prediction isn't goal



# Samples

- retail & marketing: identify buying patterns / recommendation systems
- banking: fraud detection / identify clusters (loyal, churn, ...)
- insurance: fraud dectection / risk
- publication: auto-categorize / recommend
- medicine: characterize behavior
- biology: group genes / cluster genetic markers (family ties)

# Where

- exploratory data analysis
- summary generation
- outlier detection
- finding duplicates
- pre-processing step

# Algorithms 

### **partitioned-based**
##### methods: K-means, K-median, Fuzzy, c-means, ....
##### des: sphere like clusters 
##### where: medium or large data
<br/>



### **Hierarchical**
##### methods: agglomerative, divisive
##### des: trees of clusters 
##### where: small size datasets
<br/>



### **density-based**
##### method(s): DBSCAN 
##### des: arbitrary shaped
##### where: good for special clusters or noisy data
<br/>
