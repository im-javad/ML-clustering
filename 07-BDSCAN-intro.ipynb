# Intro

Popular and widely used algorithm in clustering <br/>
**Spherical Shape Clustering**: K-Means will assign every datapoint to a cluster, no outlier <br/>
**Arbitrary Shape Clusters**: Density Based (DB) clusters will find dense areas and will seprate outliers. <br/>
Good for anomaly detection <br/>
Density: number of points within a radius <br/>



DBSCAN algorithms is effective for tasks like class identification <br/>
Effective even in presence of noise <br/>
Grouping same weather on dense areas <br/>



DBSCAN: Density Based Spatial Clustering of Applications with Noise(unlike the k-means, which was sensitive ti noise for clustering)
  - is one of the most common clustering algorithms
  - works based on density of objects

R: Radius on neighborhood 
  - Radius(R) that if includes enough number of points within, we call ot dense area

M: Min number of neighbors
  - the minimum number of data points we want in a neighborhood to define a cluster 

# Points Types

### Core:
within our neighborhood of the point there are at least M points

### Border:
 - less than M in neighborhood
 - reachable from a core point

### Outlier
outlier isn't core neigher a border 

# Advantages

- Arbitrarily shaped clusters
- Robust to outliers
