# How does it cluster?

***Iterative Mechanism***

- decide the number of cluster => K
- init K "centroids" by:
     - random points from the dataset
     - random points
- assign each customer to the closest cetroid and create distance matrix
- SSE = n1 (Xi - Cj)^2
- update the centroid to the mean of its datapoints
- continue till the centroids stop moving
- notes:
     - iterative
     - does not guarantee the best result. may catch a local optimum but its fast, so we can run it many times!  

# How can we evaluate?

* External: compare with truth
* Internal: average distance between datapoints within a cluster or the distance between cluster
* choosing K is difficult so we run with diffrent Ks and check the accuracy (say mean mean distance inside a cluster) BUT decreasing K will always reduces this. so we do the elbow method
* too many K also makes our algorithm meaningless, and we need to choose an optimal K.
